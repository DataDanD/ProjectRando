```python
# Make max dictinary with [Name, Nationality, Overall, Preferred Positions]

maxrows, maxname = [], []
for x in list(fdf.columns):
    row = fdf[x].argmax() 
    maxrows.append(row)
    maxname.append(x)
      
x = fdf.max()
maxvals = list(x)

maxmen = []
for x, y in enumerate(maxrows):
    maxmen.append(df.iloc[y][:8])

maxing = pd.DataFrame()
for x in range(len(maxmen)):
    maxing = maxing.append(maxmen[x])
maxing['id'] = maxing['Unnamed: 0']
maxin = maxing.drop('Unnamed: 0',axis=1)

maxer = pd.DataFrame([maxrows, maxname, maxvals])
x = pd.DataFrame.transpose(maxer)
x['id'] = x[0]
x = x.drop(0,axis=1)

y = pd.merge(x, maxin, left_on='id', right_on='id', how='inner')
y = y.drop_duplicates(subset=None, keep='first', inplace=False)
y.columns = ['Max Trait', 'Max', 'id', 'Age', 'Club', 'Name', 
             'Nationality', 'Overall', 'Potential', 'Value']

yy = y.drop('id',axis=1).sort_values('Club', ascending=False).sort_values('Max', ascending=False)
yy[['Max Trait', 'Max', 'Overall', 'Potential', 'Name', 'Age', 'Value', 'Club', 'Nationality',]].reset_index(drop=True)
```
